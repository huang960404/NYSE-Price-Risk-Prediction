```python
# Define exponential weights
def get_exp_weights(halftime, trail = 252):
    lamb = 2**(-1/halftime)
    wei = [lamb**i for i in range(1,trail+1)]
    return wei
weights = get_exp_weights(31, 126)
```

```python
def get_momentum(cur_date, sym):
    print(cur_date, sym)
    result = 0
    cur_close = float(prices[(prices['date']==cur_date) & (prices['symbol']==sym)]['close'])
    for i in range(126):
        count = 0
        prev_date = str((pd.to_datetime(cur_date) - pd.offsets.Day(1)))[:10]
        # while (prices[(prices['date']==prev_date) & (prices['symbol']==sym)].empty) or (rmrf[(rmrf['date']==prev_date)].empty):
        while (prices[(prices['date']==prev_date) & (prices['symbol']==sym)].empty):
        # print("jump", prev_date)
            prev_date = str((pd.to_datetime(prev_date) - pd.offsets.Day(1)))[:10]
        # cur_close = float(prices[(prices['date']==cur_date) & (prices['symbol']==sym)]['close'])
        prev_close = float(prices[(prices['date']==prev_date) & (prices['symbol']==sym)]['close'])
        return_rate = (cur_close-prev_close)/prev_close
        # rm = float(rmrf[rmrf['date'] == cur_date]['rm'])
        rf = rmrf[rmrf['date'] == cur_date]['rf']
        rf = float(rf)
        result += weights[i]*(np.log(1+return_rate)-np.log(1+rf))
        cur_date = prev_date
        cur_close = prev_close
    return result
```

### Liquidity--STOM

```python
fund = pd.read_csv('factors_temp_momentum.csv')
prices = pd.read_csv("data/prices_adjusted_clean.csv")
```

```python
fund.head()
```

```python
prices.head()
```

```python
def get_STOM(cur_date, sym, eso):
    # print(cur_date, sym)
    temp_df = prices[prices['symbol']==sym]
    temp_df = temp_df.reset_index()
    end_date = int(temp_df[temp_df['date'] == cur_date].index.values)
    end_date-21
    temp_df = temp_df.iloc[end_date-20:end_date+1]
    return np.log(sum(temp_df['volume'])/eso)
```

```python
fund['Liq-STOM'] = fund.apply(lambda row: get_STOM(row['Period Ending'], row['Ticker Symbol'], row['Estimated Shares Outstanding']), axis=1)
```

### Liquidity--STOQ

```python
def get_STOQ(cur_date, sym, eso):
    # print(cur_date, sym)
    temp_df = prices[prices['symbol']==sym]
    temp_df = temp_df.reset_index()
    end_date = int(temp_df[temp_df['date'] == cur_date].index.values)
    temp_df1 = temp_df.iloc[end_date-20:end_date+1]
    temp_df2 = temp_df.iloc[end_date-41:end_date-20]
    temp_df3 = temp_df.iloc[end_date-62:end_date-41]
    result = np.log((sum(temp_df1['volume']) + sum(temp_df2['volume']) + sum(temp_df3['volume']))/(3*eso))
    return result
```

```python
fund['Liq-STOQ'] = fund.apply(lambda row: get_STOQ(row['Period Ending'], row['Ticker Symbol'], row['Estimated Shares Outstanding']), axis=1)
```

```python
print(sum(fund['Liq-STOQ'].isnull()))
fund['Liq-STOQ'].describe()
```

```python
fund.head()
```

### Liquidity--STOH

```python
def get_STOH(cur_date, sym, eso):
    # print(cur_date, sym)
    temp_df = prices[prices['symbol']==sym]
    temp_df = temp_df.reset_index()
    end_date = int(temp_df[temp_df['date'] == cur_date].index.values)
    temp_df1 = temp_df.iloc[end_date-20:end_date+1]
    temp_df2 = temp_df.iloc[end_date-41:end_date-20]
    temp_df3 = temp_df.iloc[end_date-62:end_date-41]
    temp_df4 = temp_df.iloc[end_date-83:end_date-62]
    temp_df5 = temp_df.iloc[end_date-104:end_date-83]
    temp_df6 = temp_df.iloc[end_date-125:end_date-104]
    result = np.log((sum(temp_df1['volume']) + sum(temp_df2['volume']) + sum(temp_df3['volume']) + sum(temp_df4['volume']) + sum(temp_df5['volume']) + sum(temp_df6['volume']))/(6*eso))
    return result
```

```python
fund['Liq-STOH'] = fund.apply(lambda row: get_STOH(row['Period Ending'], row['Ticker Symbol'], row['Estimated Shares Outstanding']), axis=1)
```
