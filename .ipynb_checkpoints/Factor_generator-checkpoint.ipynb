```python
def date_transfer(day):
    return day[0:10]
prices['date'] = prices.apply(lambda x: date_transfer(x['date']), axis=1)
prices.to_csv("prices_clean.csv", index = False)
fund.to_csv("fundamentals_clean.csv", index = False)
```
